```python
from scipy.signal import find_peaks 
peaks, prop = find_peaks(t2.data, height = 0.2, distance = 400, prominence = 0.2, width = 2)
plot (t, t2.data)
y = ones(size(peaks))
plot(t[peaks], y, '*')
xlim([58, 68])
ylim([-1.1, 1.1])
show()
```
